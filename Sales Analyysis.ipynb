# Exploratory Data Analysis

##### Gender

```python
# Gender wise split 
ax = sns.countplot(x='Gender', data=df)

# Add labels to the bars
for bar in ax.containers:
    ax.bar_label(bar)

# Show the plot
plt.show()
```

```python
# total gender vs amount
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y = 'Amount',data=sales_gen)

plt.show()
```

###### From above grpahs we can see that most of the byers are female and even purchasing power of women are more than men

#### Age

```python
# Age wise split of each gender
ax = sns.countplot(x='Age Group', hue = 'Gender',data=df)

# Add labels to the bars
for bar in ax.containers:
    ax.bar_label(bar)

# Show the plot
plt.show()
```

```python
# total amount vs age group
sales_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y = 'Amount',data=sales_age)

plt.show()
```

###### From above data we can easily clasify that most of the buyers are feamle with the age group of 26-35

#### State

```python
#total numbers of orders top 10 states
sales_state =df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y = 'Orders',data=sales_state)

plt.show()
```

```python
#total amount/sales top 10 states
sales_state =df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(15)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y = 'Amount',data=sales_state)

plt.show()
```

#### From above we can say that most of the orders are fronm UP/Maharshtra and Karnataka whether it is sales quantity or amount

##### Marital_Status

```python
 ax = sns.countplot(x= 'Marital_Status', data = df)
for bar in ax.containers:
    ax.bar_label(bar)

plt.show()
    
```

```python
#total amount/sales top 10 states
sales_state =df.groupby(['Gender', 'Marital_Status'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status',y = 'Amount',hue = 'Gender',data=sales_state)

plt.show()
```

##### Most of the buyers are married women with hoigher purchasing power

#### Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(x= 'Occupation', data = df)
for bar in ax.containers:
    ax.bar_label(bar)

plt.show()
    
```

```python
#total amount/sales top 10 states
sales_state =df.groupby(['Gender', 'Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation',y = 'Amount',hue = 'Gender',data=sales_state)

plt.show()
```

###### Most of the buyers are women working at IT sector/ Healthcar & Aviation

##### Product Category

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(x= 'Product_Category', data = df)
for bar in ax.containers:
    ax.bar_label(bar)

plt.show()
    
```

```python
#total amount/sales top 10 states
sales_state =df.groupby(['Gender', 'Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_Category',y = 'Amount',hue = 'Gender',data=sales_state)

plt.show()
```

###### Most of the products that are being sold are Food/Footwear and clothing and Electronics

```python
#total amount/sales top 10 states
sales_state =df.groupby(['Gender', 'Product_ID'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_ID',y = 'Amount',hue = 'Gender',data=sales_state)

plt.show()
```
